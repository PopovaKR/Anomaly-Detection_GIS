```python
import lasio
from sys import stdout
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
las = lasio.read(r"/content/ZK_1.las")
Keys=las.keys()
print (Keys)
```

```python
gk= las['GK']
ps= las['PS']
dept = las['DEPT']
# facie = las['FACIE']
print(dept[0])
dept_0 = dept[0]
gk1 = []
dept1= []
ps1 = []
gk2 = []
dept2= []
ps2 = []
t_1 = 2850.8
b_1 = 2880.8
b_2 = 2906.0
```

```python
for i in dept:
  if (i >= t_1 and i <= b_1 ):
    dept1.append(i)
    n = int((i - dept_0)*10)
    # print(n)
    gkk1 = gk[n]
    pss1 = ps[n]
    gk1.append(gkk1)
    ps1.append(pss1)
  if (i >= b_1 and i <= b_2):
    dept2.append(i)
    k = int((i -  dept_0)*10)
    # print(n)
    gkk2 = gk[k]
    pss2 = ps[k]
    gk2.append(gkk2)
    ps2.append(pss2)
# print(dept2)
# print(gk2)
# print(ps2)
# print(dept1)
# print(gk1)
# print(ps1)
```

```python
l = 1600
df = pd.DataFrame({'DEPT': dept1,
 'PS': ps1})
df.to_csv (f'{l}P_U2_PS.csv', index = False )
df = pd.DataFrame({'DEPT': dept1,
 'GK': gk1})
df.to_csv (f'{l}P_U2_GK.csv', index = False )
df = pd.DataFrame({'DEPT': dept2,
 'PS': ps2})
df.to_csv (f'{l}P_U3_PS.csv', index = False )
df = pd.DataFrame({'DEPT': dept2,
 'GK': gk2})
df.to_csv (f'{l}P_U3_GK.csv', index = False )
files.download(f'{l}P_U2_PS.csv')
files.download(f'{l}P_U2_GK.csv')
files.download(f'{l}P_U3_PS.csv')
files.download(f'{l}P_U3_GK.csv')
```

```python
fig, ax1 = plt.subplots(figsize=(12, 4))

color = 'tab:red'
ax1.set_xlabel("Depth, m")
ax1.set_ylabel("GK, gAPI", color=color)
ax1.plot( dept2, gk2, color=color, label=str(Keys[1]))
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel("PS, mV", color=color)
ax2.plot( dept2, ps2, color=color, label=u'NGL')
ax2.tick_params(axis='y', labelcolor=color)

plt.grid(True)
#plt.legend(loc='upper right')

fig.tight_layout() 
#сохранение изображения
fig.savefig(f'{l}P_U_3.png', dpi=200, orientation='landscape')
# files.download(f'{l}P_U_3.png')
plt.show()
```
